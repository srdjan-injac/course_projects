```python
#now lets calculate most number of matches and the win percentage 
```

```python
df_match.columns
```

```python
df_match[['team1','team2']]
```

```python
matches_played = df_match['team1'].value_counts() + df_match['team2'].value_counts()
matches_played
```

```python
df_played_matches = matches_played.reset_index()
```

```python
df_played_matches.columns = ['teams','matches_played']
```

```python
df_played_matches
```

```python
df_wins = pd.DataFrame(df_match['winner'].value_counts().reset_index())
```

```python
df_wins.columns = ['teams','win_count']
```

```python
df_final_merged = pd.merge(df_played_matches,df_wins,on='teams',how='inner')
df_final_merged.sort_values(by='win_count',ascending= False)
```

```python
df_final_merged['win_percentage'] = df_final_merged['win_count']/df_final_merged['matches_played']*100
```

```python
df_final_merged.sort_values(by='win_percentage',ascending=False)
```

```python
df_melted = df_final_merged.melt(id_vars=['teams'], value_vars=['matches_played', 'win_percentage'], 
                    var_name='Metric', value_name='Value')

plt.figure(figsize=(11, 8),dpi=320)
custom_palette = {'matches_played': 'steelblue', 'win_percentage': 'orange'}
barplot = sns.barplot(data=df_melted, x='teams', y='Value', hue='Metric', palette=custom_palette)
for bar, (metric, value) in zip(barplot.patches, zip(df_melted['Metric'], df_melted['Value'])):
    x = bar.get_x() + bar.get_width() / 2 
    y = bar.get_height()  
    text = f'{value:.1f}%' if metric == 'win_percentage' else f'{int(value)}'
    barplot.text(x, y, text, ha='center', va='bottom', fontsize=8, color='black')
barplot.set_xticks(barplot.get_xticks())
barplot.set_xticklabels(df_final_merged['teams'], rotation=45, ha='right')
plt.title('Matches Played and Winning Percentage', fontsize=14)
plt.ylabel('Value')
plt.xlabel('Teams')
plt.legend(title='Metrics', loc='upper right', bbox_to_anchor=(1.25, 1)) 


plt.tight_layout()
```
